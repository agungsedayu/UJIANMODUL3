## Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel_bookings.csv')
df
```

```python
hotel = df.iloc[0:2999]
hotel
```

```python
hotel.isna().sum()
```

```python
kolom_kosong = (hotel.isnull().sum())
print(kolom_kosong[kolom_kosong > 0])
```

## Mengisi NaN
### country
NaN pada country dengan mean tidak saya buang kareana ini masih mungkin diisi dan hanya cuma satu

```python
hotel['country'].value_counts()
```

```python
hotel['country'] = hotel['country'].fillna('PRT')
```

### Agent
mengisi NaN pada kolom agent dengan 0, karena kolom ini typenya kategorical berarti identik dengan perusahaan travel agencynya jadi apabila kita isi dengan median akan ketarik pada travel agent yg median serta akan lebih fair jika kita isi dengan 0

```python
hotel['agent'].value_counts()
```

```python
hotel['agent'] = hotel['agent'].fillna(0)
```

### Company
membuang kolom company karena >90% missing atau NaN, jadi bingung harus diisi dengan apa

```python
hotel = hotel.drop(['company'], axis=1)
```

## C. Model Building & Hyper-parameter Tuning

### splitting model

```python
hotel_ML = hotel[['required_car_parking_spaces', 'lead_time', 'booking_changes','is_canceled']].copy()
hotel_ML.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
x = hotel_ML.drop(['is_canceled'], axis=1)
y = hotel_ML['is_canceled']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, shuffle=False)
```

### fitting model

```python
model_LogReg = LogisticRegression()
model_LogReg.fit(x_train, y_train)
print(model_LogReg.coef_)
print(model_LogReg.intercept_)

m = model_LogReg.coef_[0][0]
c = model_LogReg.intercept_[0]
```

```python
# prediksi
y_pred = model_LogReg.predict(x_test)
```

## model performence

```python
model_LogReg.score(x_train, y_train)
```

```python
model_LogReg.score(x_test, y_test)
```

```python
# Parameter yang dipakai di model asli
model_LogReg.get_params()
```

```python
# parameter model linear regression yang akan dituned + nilai yang mungkin

penalty = ['l1', 'l2', 'elasticnet', 'none']
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
max_iter = [1, 10, 100, 1000, 10000]

param = {'penalty': penalty, 'solver': solver, 'max_iter': max_iter}
param
```

### randomized search cv

```python
# mencari parameter terbaik pada model: logistic regression

from sklearn.model_selection import RandomizedSearchCV
model_LR = LogisticRegression()
model_LR_RS = RandomizedSearchCV(
    estimator = model_LR, param_distributions= param, cv = 5
)
```

```python
model_LR_RS.fit(x_train, y_train)
model_LR_RS.best_params_
```

```python
model_LogReg.score(x_test, y_test)
```

```python
model_LogReg_Baru = LogisticRegression(solver='saga', penalty = 'l2', max_iter = 10000)

model_LogReg_Baru.fit(x_train, y_train)
model_LogReg_Baru.score(x_test, y_test)
```

## grid search cv

```python
from sklearn.model_selection import GridSearchCV
model_LR2 = LogisticRegression()
model_LR2_GS = GridSearchCV(
    model_LR2, param, cv = 5
)
```

```python
model_LR2_GS.fit(x_train, y_train)
model_LR2_GS.best_params_
```

```python
model_LogReg.score(x_test, y_test)
```

```python
model_LogReg_Baru_2 = LogisticRegression(solver='lbfgs', penalty = 'none', max_iter = 10000)

model_LogReg_Baru_2.fit(x_train, y_train)
model_LogReg_Baru_2.score(x_test, y_test)
```
